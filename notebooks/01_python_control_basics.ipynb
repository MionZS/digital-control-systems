# 01 — python-control basics

This notebook demonstrates creating LTI models, computing step responses and Bode plots.

```python
import numpy as np
import matplotlib.pyplot as plt
import control

from control_lab.models.lti import LTIModel
from control_lab.utils.plotting import plot_bode

%matplotlib inline
```

## Create a mass-spring-damper model

```python
model = LTIModel.mass_spring_damper(m=1.0, k=1.0, c=0.5)
print('A =', model.A)
print('B =', model.B)
print('C =', model.C)
print('Poles:', model.poles())
print('Is stable:', model.is_stable())
```

## Step response

```python
sys = control.ss(model.A, model.B, model.C, model.D)
t, y = control.step_response(sys)

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t, y, linewidth=2)
ax.set_xlabel('Time (s)')
ax.set_ylabel('Position')
ax.set_title('MSD Step Response')
ax.grid(True)
plt.tight_layout()
plt.show()
```

## Bode plot

```python
fig = plot_bode(model, title='MSD Bode Plot')
plt.show()
```
